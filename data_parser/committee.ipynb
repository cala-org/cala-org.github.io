#### Insert comm-chair and comm-member data

```python
#node-2182 > div.content.clearfix > div.field.field-name-body.field-type-text-with-summary.field-label-hidden > div > div > ul:nth-child(31)
#node-2182 > div.content.clearfix > div.field.field-name-body.field-type-text-with-summary.field-label-hidden > div > div > ul:nth-child(37)
#node-2182 > div.content.clearfix > div.field.field-name-body.field-type-text-with-summary.field-label-hidden > div > div > ul:nth-child(41)
ul_number = 53
ul_selector = "#node-" + node_id + " > div.content.clearfix > div.field.field-name-body.field-type-text-with-summary.field-label-hidden > div > div > ul:nth-child(" + str(ul_number) + ")"
result = soup.select(ul_selector)


#node-2182 > div.content.clearfix > div.field.field-name-body.field-type-text-with-summary.field-label-hidden > div > div > ul:nth-child(35) > li:nth-child(4) > ul
# alter_ul_selector = "#node-" + node_id + " > div.content.clearfix > div.field.field-name-body.field-type-text-with-summary.field-label-hidden > div > div > ul:nth-child(35) > li:nth-child(4) > ul"
# result = soup.select(alter_ul_selector)

resArr = []
for item in result:
    resStr = item.get_text().replace("\u200b","")
    resArr.append(resStr)

resArrSplit = resArr[0].splitlines()
print(resArrSplit)




com_chair_cochair_liOutputArr = []
com_member_liOutputArr = []

for item in resArrSplit:
    if len(item) > 5:
        chairItemDic = {"name":"","email":""}
        itemDic = {"name":"","email":""}
        nameString = item.split(" ", 2)[0].replace('\u200b', '').replace("\u00a0","").strip() + " " + item.split(" ", 2)[1].replace('\u200b', '').replace("\u00a0","").strip()
        emailString = item.split(" ", 2)[2].replace('\u200b', '').replace("\u00a0","").replace("\xa0","").replace("(","").replace(")","").strip()
        # nameString = item.split(",")[0].replace('\u200b', '').replace("\u00a0","").strip()
        # emailString = item.split(",")[1].replace('\u200b', '').replace("\u00a0","").replace("\xa0","").replace("(","").replace(")","").strip()
        if emailString.find("Co-Chair") != -1:
            chairItemDic.update({"name": nameString, "email":emailString.replace("Co-Chair","").strip()})
            com_chair_cochair_liOutputArr.append(chairItemDic)
        elif emailString.find("Chair") != -1:
            chairItemDic.update({"name": nameString, "email":emailString.replace("Chair","").strip()})
            com_chair_cochair_liOutputArr.append(chairItemDic)
        else:
            itemDic.update({"name": nameString, "email":emailString.replace("(Member)","").strip()})
            com_member_liOutputArr.append(itemDic)

blockDict['comm-chair'] = com_chair_cochair_liOutputArr
blockDict['comm-members'] = com_member_liOutputArr


json_object = json.dumps(blockDict, indent = 4) 
print(json_object)

# print(blockDict)
```
